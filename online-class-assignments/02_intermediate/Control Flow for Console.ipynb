# High-Low Game: Python Control Flow Practice

This notebook implements a simple High-Low game where the user guesses if their randomly generated number is higher or lower than the computer's number. The game runs for multiple rounds, and the player receives points for correct guesses. The game logic includes input validation and displays performance-based messages at the end

```python
import random

NUM_ROUNDS = 5

def main():
    print("Welcome to the High-Low Game!")
    print('--------------------------------')

    # Variable to keep track of score
    your_score = 0

    # Loop through the rounds
    for i in range(NUM_ROUNDS):
        print("Round", i + 1)

        # Generate random numbers for both the user and the computer
        computer_num = random.randint(1, 100)
        your_num = random.randint(1, 100)
        print("Your number is", your_num)

        # Get user's guess (higher or lower)
        choice = input("Do you think your number is higher or lower than the computer's?: ")

        # Input validation for choice
        while choice != "higher" and choice != "lower":
            choice = input("Please enter either 'higher' or 'lower': ")

        # Check if user's guess is correct
        higher_and_correct = choice == "higher" and your_num > computer_num
        lower_and_correct = choice == "lower" and your_num < computer_num

        # Update score if guess is correct
        if higher_and_correct or lower_and_correct:
            print("You were right! The computer's number was", computer_num)
            your_score += 1
        else:
            print("Aww, that's incorrect. The computer's number was", computer_num)

        # Show current score
        print("Your score is now", your_score)
        print()

    # Show final score and performance message
    print("Your final score is", your_score)

    if your_score == NUM_ROUNDS:
        print("Wow! You played perfectly!")
    elif your_score > NUM_ROUNDS // 2:
        print("Good job, you played really well!")
    else:
        print("Better luck next time!")

if __name__ == "__main__":
    main()
```
