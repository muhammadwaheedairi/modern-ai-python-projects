# ðŸŽ¯ Number Guessing Game (User guesses computer's number)

**Description**:
The computer picks a random number between 1 and 100. You guess the number. The computer tells you if your guess is too high, too low, or correct.

**How to Run**:

Run the code.

Enter a number between 1 and 100.

Keep guessing until you find the correct number.



```python
import random

# Welcome message
print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 100.")
print("Can you guess what it is?\n")

# Computer selects a random number between 1 and 100
secret_number = random.randint(1, 100)

# Initialize guess variable
guess = None

# Keep asking until the user guesses correctly
while guess != secret_number:
    try:
        guess = int(input("Enter your guess: "))

        if guess < secret_number:
            print("Too low! Try again.\n")
        elif guess > secret_number:
            print("Too high! Try again.\n")
        else:
            print(f"Congratulations! You guessed it right. The number was {secret_number}!")
    except ValueError:
        print("Please enter a valid number.\n")

print("\nThanks for playing! ðŸŽ‰")
```
