# Mad Libs Story Maker

## Description
This is a fun Mad Libs game where the user provides random words (noun, verb, adjective, place), and the program creates a funny story using those words.

## How to Run
- Run the code.
- Enter:
  - a noun
  - a verb
  - an adjective
  - a place
- Read and enjoy the story!

---


```python
# Mad Libs Game
print("Welcome to the Mad Libs game!")

# Collect words from the user
noun = input("Enter a noun: ")
verb = input("Enter a verb: ")
adjective = input("Enter an adjective: ")
place = input("Enter a place: ")

# Create the story
story = f"One day, a {adjective} {noun} decided to {verb} at the {place}. It was an unforgettable adventure!"

# Display the story
print("\nHere is your Mad Libs story:")
print(story)
```
