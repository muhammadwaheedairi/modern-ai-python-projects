## Description:
Play Rock, Paper, Scissors against the computer. The computer randomly picks rock, paper, or scissors, and you try to beat it.

## How to Run:

Run the code.

Enter your choice: rock, paper, or scissors.

See if you win, lose, or tie.



```python
import random

# Welcome message
print("Welcome to Rock, Paper, Scissors Game!")
print("You will be playing against the computer.")
print("Make your move by typing: rock, paper, or scissors.\n")

# Available choices
choices = ['rock', 'paper', 'scissors']

# Computer makes a random choice
computer_choice = random.choice(choices)

# User input
user_choice = input("Enter your choice (rock, paper, scissors): ").lower()

# Check if the user's choice is valid
if user_choice not in choices:
    print("\nInvalid choice. Please select rock, paper, or scissors.")
else:
    print(f"\nComputer chose: {computer_choice}")

    # Determine the winner
    if user_choice == computer_choice:
        print("It's a tie!")
    elif (
        (user_choice == 'rock' and computer_choice == 'scissors') or
        (user_choice == 'paper' and computer_choice == 'rock') or
        (user_choice == 'scissors' and computer_choice == 'paper')
    ):
        print("Congratulations! You win!")
    else:
        print("Oops! Computer wins!")

print("\nThanks for playing! ðŸ‘‹")
```
