# ðŸ¤– Reverse Number Guessing Game (Computer guesses your number)

**Description**:
You think of a number between 1 and 100. The computer will guess it using smart binary search logic. You tell if its guess is too high, too low, or correct.

**How to Run:**

Run the code.

Think of a number silently between 1 and 100.

After each guess, type:

h (high) if guess is too high

l (low) if guess is too low

c (correct) if computer guessed correctly



```python
print("Welcome to the Reverse Number Guessing Game!")
print("Think of a number between 1 and 100, and I will try to guess it.")
print("After each guess, tell me if I'm too high (h), too low (l), or correct (c).\n")

low = 1
high = 100
feedback = ''

while feedback != 'c':
    if low > high:
        print("Hmm... something doesn't add up. Are you sure you gave the right feedback?")
        break

    guess = (low + high) // 2
    print(f"My guess is: {guess}")
    feedback = input("Is it too high (h), too low (l), or correct (c)? ").lower()

    if feedback == 'h':
        high = guess - 1
    elif feedback == 'l':
        low = guess + 1
    elif feedback == 'c':
        print(f"\nYay! I guessed your number {guess} correctly! ðŸŽ‰")
    else:
        print("Please enter 'h' for high, 'l' for low, or 'c' for correct.")

print("\nThanks for playing! ðŸ¤–")
```
