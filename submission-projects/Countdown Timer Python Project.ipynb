## Description:
Create a simple countdown timer that counts down seconds and displays remaining time every second.

## How to Run:

Run the code.

Enter the number of seconds for countdown.

Watch the countdown, and at 0 seconds, see "Time's up!".



```python
import time

# Welcome message
print("Welcome to the Countdown Timer!")
print("Enter the number of seconds for the countdown.\n")

# Get user input for the countdown time
try:
    seconds = int(input("Enter the number of seconds: "))

    if seconds <= 0:
        print("Please enter a positive number greater than 0.")
    else:
        # Countdown loop
        while seconds > 0:
            print(f"Time remaining: {seconds} seconds")
            time.sleep(1)  # Wait for 1 second
            seconds -= 1  # Decrease the remaining time

        # When the countdown finishes
        print("Time's up! üéâ")
except ValueError:
    print("Please enter a valid number.")
```
