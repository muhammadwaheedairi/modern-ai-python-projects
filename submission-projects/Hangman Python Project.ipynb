 # ðŸ§µ Hangman Game

## Description:
Try to guess a hidden word one letter at a time. You have limited attempts. Wrong guesses are tracked separately.

## How to Run:

Run the code.

Guess one letter at a time.

Try to complete the word before running out of attempts.



```python
import random

# Welcome message
print("Welcome to Hangman Game!")
print("Guess the word one letter at a time.\n")

# List of words to choose from
word_list = ['python', 'programming', 'hangman', 'challenge', 'computer']

# Select a random word
word = random.choice(word_list)
word_letters = set(word)  # Unique letters in the word
guessed_letters = set()   # Correct letters guessed
wrong_guesses = set()     # Incorrect letters guessed
attempts = 6  # Number of allowed wrong guesses

# Game loop
while len(word_letters) > 0 and attempts > 0:
    # Display current progress
    current_word = [letter if letter in guessed_letters else '_' for letter in word]
    print('Word:', ' '.join(current_word))
    print(f"Wrong guesses: {', '.join(wrong_guesses)}")
    print(f"Attempts left: {attempts}\n")

    # User guesses a letter
    guess = input("Guess a letter: ").lower()

    # Validating user input
    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a single valid letter.\n")
        continue

    # Checking the guessed letter
    if guess in word_letters:
        guessed_letters.add(guess)
        word_letters.remove(guess)
        print(f"Good job! '{guess}' is in the word.\n")
    elif guess in guessed_letters or guess in wrong_guesses:
        print("You already guessed that letter. Try a different one.\n")
    else:
        wrong_guesses.add(guess)
        attempts -= 1
        print(f"Oops! '{guess}' is not in the word.\n")

# Game result
if len(word_letters) == 0:
    print(f"Congratulations! You guessed the word '{word}' correctly! ðŸŽ‰")
else:
    print(f"Game Over! The word was '{word}'. Better luck next time!")

print("\nThanks for playing Hangman! ðŸŽ¯")
```
