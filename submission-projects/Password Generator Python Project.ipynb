## Description:
Generate a strong random password with letters, numbers, and symbols based on the user-specified length.

## How to Run:

Run the code.

Enter the length of password you want (preferably 8 or more).

Copy the generated secure password.



```python
import random
import string

# Welcome message
print("Welcome to the Password Generator!")
print("You can generate a secure password using letters, numbers, and symbols.\n")

# Ask the user for the desired password length
try:
    length = int(input("Enter the desired password length: "))

    if length < 8:
        print("For security reasons, it's recommended to have a password length of at least 8 characters.")
    else:
        # Define the character pool (letters, digits, and symbols)
        all_characters = string.ascii_letters + string.digits + string.punctuation

        # Generate a random password
        password = ''.join(random.choice(all_characters) for _ in range(length))

        # Display the generated password
        print(f"Your generated password is: {password}\n")

except ValueError:
    print("Please enter a valid number for the password length.")
```
